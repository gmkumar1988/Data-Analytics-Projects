```python
import numpy as np
diabetes_dataset = diabetes_dataset.replace(np.nan,99999)
```

```python
diabetes_dataset.columns
```

```python
print("Splitting Data: ")
X,y = diabetes_dataset[['age','height','weight']].values,diabetes_dataset['chol'].values
```

Splitting into Train and Test Set data: 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=30, random_state=0)
```

Train Model using Logistic Regression (Classification Type)

```python
from sklearn.linear_model import LogisticRegression
print("Training model:")
model = LogisticRegression(multi_class='ovr',C=1/0.1, solver="liblinear").fit(X_train,y_train)
```

Evaluate the model: 

```python
import numpy as np
y_hat = model.predict(X_test)
acc = np.average(y_hat == y_test)
print("Accuracy value is :",acc)
```

```python
from sklearn.metrics import roc_auc_score
y_scores = model.predict_proba(X_test)
# auc_score = roc_auc_score(y_test,y_scores)
```
