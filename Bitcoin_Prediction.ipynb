```python
features = ['open','high','low','close']
```

```python
plt.subplots(figsize = (15,5))
for i, col in enumerate(features):
    plt.plot(2,2,i+1)
    sb.displot(bitcoin_dataset[col])
plt.show()
```

```python
plt.subplots(figsize = (12,5))
for i, col in enumerate(features):
    plt.subplot(2,2, i+1)
    sb.boxplot(bitcoin_dataset[col])
plt.show()
```

Feature Engineering: 

```python
splitted = bitcoin_dataset['date'].str.split('-',expand=True)
```

```python
bitcoin_dataset['year'] = splitted[0].astype('int')
bitcoin_dataset['month'] = splitted[1].astype('int')
```

```python
bitcoin_dataset['date'] = pd.to_datetime(bitcoin_dataset['date'])
```

```python
data_grouped = bitcoin_dataset.groupby('year')
```
